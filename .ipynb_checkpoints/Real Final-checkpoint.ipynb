# Introduction
This project takes a dataset of Indian undergraduate students' responses describing a student's usage of AI and variables like their major (stream), whether professors allow them to use AI, and the impact they believe their AI usage has on their grade. I intend to train machine learning models on this data to determine which variables have the greatest impact on a student's grade, and in what direction these variables have an impact.

I hypothesize that use cases which are more focused on learning or studying instead of assignment completion will positively contribute to the perceived impact on grades. I also hypothesize that more daily usage hours of AI tools will do the same thing, because long sessions with AI represent more of a conversation than rote task completion.

# Data Processing

```python
import pandas as pd
df = pd.read_csv('Students.csv')
print(df.head())
```

As we can see, the dataset has many non-numeric variables, and an ML model needs numeric variables to train on, so the data must be processed.
## Irrelevant Variables
Some variables don't have any relevance to predictions that I care to measure and should be removed. This includes Student_Name, College_Name, and State.

```python
irrelevant_cols = ['Student_Name', 'College_Name', 'State']
df = df.drop(columns=irrelevant_cols)
```

## Binary Variables
The yes/no variables can just be converted into 0s and 1s.

```python
binary_columns = ["Do_Professors_Allow_Use", "Willing_to_Pay_for_Access"]
df[binary_columns] = df[binary_columns].replace({
    'Yes': 1, 'No': 0,
})
```

## Multiple-Select Variables
For some variables, respondents have multiple values, such as with Use_Cases. I believe the best and simplest way to encode this is by turning every element of the multiple select into its own binary variable, such as "Use_Assignments" if Assignments is listed as a use case.

```python
# Doing .str.get_dummies.add_prefix with a join creates multiple new columns based on whether something was found in the old one
multi_col_1 = 'Use_Cases'
df = df.join(df[multi_col_1].str.get_dummies(sep=', ').add_prefix(f'Use_'))

multi_col_2 = 'AI_Tools_Used'
df = df.join(df[multi_col_2].str.get_dummies(sep=', ').add_prefix(f'AI_Tool_'))

# The data had some repeat values, like 'Use_Exam Prep' and 'Use_Exam Preparation', which could be combined
ms_combine_1 = ['Use_Exam Prep', 'Use_Exam Preparation']
ms_combine_2 = ['Use_Project Work', 'Use_Projects']
ms_combine_3 = ['AI_Tool_Bard', 'AI_Tool_Gemini'] # these are literally the same thing
# These lines combine the values of the ms_columns, then apply a boolean to normalize any values 1 or higher to true, then apply int to encode t/f as 0 or 1
df['Use_Exam Studying'] = (df[ms_combine_1].sum(axis=1) > 0).astype(int)
df['Use_Project'] = (df[ms_combine_2].sum(axis=1) > 0).astype(int)
df['AI_Tool_Google'] = (df[ms_combine_3].sum(axis=1) > 0).astype(int)

# drop the old columns, as all their data is encoded in new ones now
df = df.drop(columns=[multi_col_1, multi_col_2])
df = df.drop(columns=ms_combine_1)
df = df.drop(columns=ms_combine_2)
df = df.drop(columns=ms_combine_3)
```

## Categorical Variables
Categorical variables (or single-select variables) such as "Device Used" can be one-hot encoded similarly to letters, since the order doesn't matter.

Before one-hot encoding though, I want to encode "Internet Access: Poor/Medium/High" as 1/2/3, since this is a variable that exists on a continuum. This will be better for assessing linear relationships, and it also reduces the variable count.

```python
df['Internet_Access'] = df['Internet_Access'].replace({
    'Poor': 1, 'Medium': 2, "High": 3
})
```

Now we can one-hot encode the other categorical variables

```python
categoric_cols = ['Stream', 'Preferred_AI_Tool', 'Device_Used']
# get dummies adds columns for all potential values in a column
df = pd.get_dummies(df, columns=categoric_cols) 
# converts new true/false of one hot encoded columsn to int
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```

## Creating Alternate Dataframes
After initial work with this data, I learned that the results are somewhat limited and drawing conclusions from those results wouldn't be simple. To work with this, creating some alternate versions of the data for error analysis is helpful. For example, I could consolidate use cases like assignment completion, content writing, and project work into a single variable called 'Use_Work', since all of these constitute a similar philosophy of a student's AI use that I'm interested to track. Another alternate dataframe I could use is only the students who list more moderate impacts on their grades, since the extreme values could be impacting predictions.
### Consolidated Dataframe
This dataframe intends to simplify the data to create a better representation for testing the conclusions I'm actually interested in. A student's preferred AI tool doesn't matter as much as what they use, and using AI for assignment completion versus content writing are very similar use cases that I don't want to differentiate between in my results. I'm also not too concerned with 'Device_Used'.

```python
# cdf stands for consolidated_df
cdf = df.copy()

# Combine similar use cases to concentrate the data
to_combine_1 = ['Use_Exam Studying','Use_MCQ Practice']
cdf['Use_Studying'] = (cdf[to_combine_1].sum(axis=1) > 0).astype(int)

to_combine_2 = ['Use_Project', 'Use_Assignments', 'Use_Content Writing', 'Use_Coding Help', 'Use_Resume Writing']
cdf['Use_Completion'] = (cdf[to_combine_2].sum(axis=1) > 0).astype(int)

to_combine_3 = ['Use_Doubt Solving', 'Use_Learning new topics', 'Use_Notes']
cdf['Use_Learning'] = (cdf[to_combine_3].sum(axis=1) > 0).astype(int)

to_combine_4 = ['AI_Tool_ChatGPT', 'AI_Tool_Google', 'AI_Tool_Copilot']
cdf['AI_Tool_Chatbot'] = (cdf[to_combine_4].sum(axis=1) > 0).astype(int)

to_combine_5 = ['AI_Tool_Claude']
cdf['AI_Tool_Coding'] = (cdf[to_combine_5].sum(axis=1) > 0).astype(int)

to_combine_6 = ['AI_Tool_Midjourney', 'AI_Tool_Other']
cdf['AI_Tool_Misc'] = (cdf[to_combine_6].sum(axis=1) > 0).astype(int)

# Drop Irrelevant Columns
preferred_ai_columns=['Preferred_AI_Tool_Bard', 'Preferred_AI_Tool_ChatGPT', 'Preferred_AI_Tool_Claude', 'Preferred_AI_Tool_Copilot', 'Preferred_AI_Tool_Gemini', 'Preferred_AI_Tool_Other']
device_used_columns=['Device_Used_Laptop', 'Device_Used_Mobile', 'Device_Used_Tablet']
to_drop = to_combine_1 + to_combine_2 + to_combine_3 + to_combine_4 + to_combine_5 + to_combine_6 + preferred_ai_columns + device_used_columns
cdf = cdf.drop(columns=to_drop)

print(list(cdf.columns))
```

### Error Analysis Dataframes
Prior testing has established that some models will not be very successful with the initial dataframe. Generating other dataframes that only focus on students with certain grade impact ranges will be very useful for seeing whether the models are more or less successful with the more extreme or more moderate grade impacts.

```python
def make_limited_range_df(df, lower_bound, upper_bound):
    mask = df['Impact_on_Grades'] >= lower_bound
    # df.loc with df condition 'mask' removes rows of the df which violate the condition
    new_df = df.loc[mask]
    mask = new_df['Impact_on_Grades'] <= upper_bound
    new_df = new_df.loc[mask]
    return new_df

# df3: only include students whose impact on grades is between -3 and 3
df3 = make_limited_range_df(df, -3, 3)
cdf3 = make_limited_range_df(cdf, -3, 3)

# df2: only include students whose impact on grades is between -2 and 2
df2 = make_limited_range_df(df, -2, 2)
cdf2 = make_limited_range_df(cdf, -2, 2)

# df1: only include students whose impact on grades is between -1 and 1
df1 = make_limited_range_df(df, -1, 1)
cdf1 = make_limited_range_df(cdf, -1, 1)

# dfp: only include students whose impact on grades is between 1 and 3 (positive impact)
dfp = make_limited_range_df(df, 1, 3)
cdfp = make_limited_range_df(cdf, 1, 3)

# dfn: only include students whose impact on grades is between -3 and -1 (negative impact)
dfn = make_limited_range_df(df, -3, -1)
cdfn = make_limited_range_df(cdf, -3, -1)
```

# Training Models
## Creating a "Mean" Model
I want to use root mean squared error as my performance metric, but it's important that I'm able to tell whether a model is actually good at predicting anything. To solve this, I'm going to use the RMSE of "predicting" the mean impact on grades for every sample as a baseline: if my model performs similarly to just guessing the mean, then the model might as well be randomly guessing. I can use the percentage the model improves over this mean prediction as a more communicative performance metric.

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# good to have an rmse function, since sklearn only provides mse
def rmse(real, pred):
    mse = mean_squared_error(real, pred)
    rmse = np.sqrt(mse)
    return rmse

# creating a function so this is easy to apply to different dataframes
# will print out relevant results for every model and df
def mean_model_test(df, model_rmse, index=0):
    mean = df['Impact_on_Grades'].mean()
    grade_impact_array = df['Impact_on_Grades'].to_numpy()
    mean_array = [mean] * len(grade_impact_array)

    mean_rmse = rmse(grade_impact_array, mean_array)
    model_improvement = -1 * ( (model_rmse - mean_rmse) / mean_rmse) * 100

    print(f"{df_titles[index]} - Model RMSE: {model_rmse:.4f}; Model Improvement: {model_improvement:.2f}%; Mean RMSE: {mean_rmse:.4f}")
```

## Linear Regression Model

### Setup
With the multi-dataframe setup, we can create a function to easily and quickly get results for all dataframes.

```python
from sklearn.linear_model import LinearRegression

def perform_linear_regression(index):
    # remember that all_Xy[index][0, 1, 2, 3] represent X_train, X_test, y_train, y_test
    reg = LinearRegression()
    reg.fit(all_Xy[index][0], all_Xy[index][2])
    pred = reg.predict(all_Xy[index][1])
    lr_rmse = rmse(all_Xy[index][3], pred)
    mean_model_test(all_dfs[index], lr_rmse, index)
```

## Random Forest Model
I'm hoping the random forest model performs better than linear regression, because the linearity in the data is very weak, and random forest models capture inherently non-linear data.
### Setup

```python
from sklearn.ensemble import RandomForestRegressor

def perform_random_forest(index):
    rf = RandomForestRegressor( # this is mostly the default configuration, and what worked best for me
        n_estimators=200,
        max_depth = None,
        min_samples_split=2,
        min_samples_leaf=1,
        max_features=1.0,
        random_state=20
    )
    rf.fit(all_Xy[index][0], all_Xy[index][2])
    rf_pred = rf.predict(all_Xy[index][1])
    rf_rmse = rmse(all_Xy[index][3], rf_pred)
    mean_model_test(all_dfs[index], rf_rmse, index)
    return rf
```

## Keras Hidden-Layer FNN
I initially tried to use a pytorch FNN, but it ended up being more volatile to hyperparameters and harder to experiment with than a Keras FNN. The pytorch process can be found in the old jupyter notebook. In that notebook, I also determined that dropout was my preferred method for preventing overfitting, but I will be experimenting with reducing the amount of hidden nodes too.
### Setup

```python
# This env variable gave me less errors
%env TF_ENABLE_ONEDNN_OPTS=0
```

```python
from tensorflow import keras
from tensorflow.keras.models import Model
# mostly taken from our lab on keras
def create_nn(n_features, n_predictions, first_nodes, second_nodes, first_dropout, second_dropout, activation='relu'):
    # Input layer
    input = keras.layers.Input(shape=(n_features,), name='input')

    # HIdden layers
    layers_dense = keras.layers.Dense(first_nodes, activation)(input)
    # The dropout rate is what % of nodes get dropped
    layers_dense = keras.layers.Dropout(first_dropout)(layers_dense)
    layers_dense = keras.layers.Dense(second_nodes, activation)(layers_dense)
    layers_dense = keras.layers.Dropout(second_dropout)(layers_dense)

    # Output layer
    output = keras.layers.Dense(n_predictions)(layers_dense)

    # Defining the model and compiling it
    return Model(inputs=input, outputs=output, name="fnn_model")
```

```python
def perform_fnn(index, first_nodes, second_nodes, first_dropout, second_dropout, epochs=200, activation='relu', verbose=0):
    n_features = all_Xy[index][0].shape[1]
    n_predictions = 1

    # passing all these parameters allows me to run it through loops and test different hyperparameters easier
    model = create_nn(n_features, n_predictions, first_nodes, second_nodes, first_dropout, second_dropout, activation=activation)
    model.compile(loss='mse', optimizer=keras.optimizers.Adam(),
                  metrics=[keras.metrics.RootMeanSquaredError()])
    
    history = model.fit(all_Xy[index][0], all_Xy[index][2],
                        batch_size = 32,
                        epochs = epochs,
                        validation_data=(all_Xy[index][1], all_Xy[index][3]), verbose = verbose)
    predictions = model.predict(all_Xy[index][1])
    fnn_rmse = rmse(all_Xy[index][3], predictions)
    mean_model_test(all_dfs[index], fnn_rmse, index)
    return model, history
```

### Training
After testing many combinations of hyperparameters, 2 layers of 40 nodes with 50% dropout and relu activation worked best at 100 epochs.

### SHAP Importance Tables
A SHAP importance value represents how much that variable impacts predictions on average. For example, a value of 0.2 for a feature means on average, that feature contributes 0.2 to the impact on grade prediction (which, for df3, is between -3 and 3).

The categorical and multi-select variables require a lot of processing to get SHAP importance scores that account for all of them.

```python
# recombine categorical or multi_select variables to measure importance
# AI Tool columns
tool_indices = [15, 17, 18, 16, 19]
# combine all the shap points for these indices
tool_shap_values = rf3_shap_values[:, tool_indices].sum(axis=1)
# take absolute values and average them for importance impact (definition of shap importance)
tool_importance = np.abs(tool_shap_values).mean()
# create a row for concatenation later
tool_row = {
    'Feature': 'Tools_Used', 
    'SHAP_Importance': tool_importance
}
# preferred AI tool columns
preferred_indices = [37, 38, 36, 34, 35, 33]
preferred_shap_values = rf3_shap_values[:, preferred_indices].sum(axis=1)
preferred_importance = np.abs(preferred_shap_values).mean()
preferred_row = {
    'Feature': 'Preferred_Tool', 
    'SHAP_Importance': preferred_importance
}
# Stream (major) columns
stream_indices = [26, 32, 24, 29, 30, 23, 31, 25, 27, 28]
stream_shap_values = rf3_shap_values[:, stream_indices].sum(axis=1)
stream_importance = np.abs(stream_shap_values).mean()
stream_row = {
    'Feature': 'Stream/Major', 
    'SHAP_Importance': stream_importance
}
# Device Used columns
device_indices = [39, 41, 40]
device_shap_values = rf3_shap_values[:, device_indices].sum(axis=1)
device_importance = np.abs(device_shap_values).mean()
device_row = {
    'Feature': 'Device_Used', 
    'SHAP_Importance': device_importance
}
# Use Columns
use_indices = [21, 9, 13, 8, 11, 7, 14, 12, 20, 10]
use_shap_values = rf3_shap_values[:, device_indices].sum(axis=1)
use_importance = np.abs(device_shap_values).mean()
use_row = {
    'Feature': 'Use_Cases', 
    'SHAP_Importance': use_importance
}

# take absolute values of all the shap points and average them (definition of shap importance)
rf3_mean_abs_shap = np.abs(rf3_shap_values).mean(axis=0)
# fill data frame with data
rf3_importance_df = pd.DataFrame({
    'Feature': rf3_X_test.columns,
    'SHAP_Importance': rf3_mean_abs_shap
})

# add combined data to dataframe
rf3_importance_df = pd.concat([rf3_importance_df, pd.DataFrame([tool_row]), pd.DataFrame([preferred_row]),pd.DataFrame([stream_row]),pd.DataFrame([device_row]),pd.DataFrame([use_row]),], ignore_index=True)
# sort data
rf3_importance_df = rf3_importance_df.sort_values(by='SHAP_Importance', ascending=False)

print(rf3_importance_df)
```

For df3, daily usage hours and major are the most important factors. It's interesting to see major play a big role, and any conclusions about the effects major has should focus back on the graph to track individual majors. However, it should be noted that 0.15 doesn't seem like a super impactful SHAP score to begin with, and this is on a model that improves less than 40% over purely guessing. On another note, Use Cases are remarkably unimportant for making predictions, being even less important than a student's major.

```python
# for crf
# recombine categorical or multi_select variables to measure importance
# AI Tool columns
tool_indices = [20, 21, 22]
tool_shap_values = crf3_shap_values[:, tool_indices].sum(axis=1)
tool_importance = np.abs(tool_shap_values).mean()
tool_row = {
    'Feature': 'Tools_Used', 
    'SHAP_Importance': tool_importance
}
# Stream (major) columns
stream_indices = [10, 16, 8, 13, 9, 11, 7, 14, 15, 12]
stream_shap_values = crf3_shap_values[:, stream_indices].sum(axis=1)
stream_importance = np.abs(stream_shap_values).mean()
stream_row = {
    'Feature': 'Stream/Major', 
    'SHAP_Importance': stream_importance
}
# Use Columns
use_indices = [18, 19, 17]
use_shap_values = rf3_shap_values[:, device_indices].sum(axis=1)
use_importance = np.abs(device_shap_values).mean()
use_row = {
    'Feature': 'Use_Cases', 
    'SHAP_Importance': use_importance
}

crf3_mean_abs_shap = np.abs(crf3_shap_values).mean(axis=0)

crf3_importance_df = pd.DataFrame({
    'Feature': crf3_X_test.columns,
    'SHAP_Importance': crf3_mean_abs_shap
})

crf3_importance_df = pd.concat([crf3_importance_df, pd.DataFrame([tool_row]), pd.DataFrame([stream_row]), pd.DataFrame([use_row])], ignore_index=True)
crf3_importance_df = crf3_importance_df.sort_values(by='SHAP_Importance', ascending=False)

print(crf3_importance_df)
```

### SHAP Importance Table

```python
# # recombine categorical or multi_select variables to measure importance
# # AI Tool columns
tool_indices = [20, 21, 22]
tool_shap_values = crf3_shap_values[:, tool_indices].sum(axis=1)
tool_importance = np.abs(tool_shap_values).mean()
tool_row = {
    'Feature': 'Tools_Used', 
    'SHAP_Importance': tool_importance
}
# Stream (major) columns
stream_indices = [10, 16, 8, 13, 9, 11, 7, 14, 15, 12]
stream_shap_values = crf3_shap_values[:, stream_indices].sum(axis=1)
stream_importance = np.abs(stream_shap_values).mean()
stream_row = {
    'Feature': 'Stream/Major', 
    'SHAP_Importance': stream_importance
}
# Use Columns
use_indices = [18, 19, 17]
use_shap_values = rf3_shap_values[:, device_indices].sum(axis=1)
use_importance = np.abs(device_shap_values).mean()
use_row = {
    'Feature': 'Use_Cases', 
    'SHAP_Importance': use_importance
}

cfnn3_mean_abs_shap = np.abs(cfnn3_shap_values).mean(axis=0)

cfnn3_importance_df = pd.DataFrame({
    'Feature': cfnn3_X_test.columns,
    'SHAP_Importance': cfnn3_mean_abs_shap
})

cfnn3_importance_df = pd.concat([cfnn3_importance_df, pd.DataFrame([tool_row]), pd.DataFrame([stream_row]), pd.DataFrame([use_row])], ignore_index=True)
cfnn3_importance_df = cfnn3_importance_df.sort_values(by='SHAP_Importance', ascending=False)

print(cfnn3_importance_df)
```

Major is the most important factor for the FNN, which speaks to its importance. Use Cases is still rather irrelevant in the predictions. 
# Conclusions
This dataset is obviously very hard for models to make predictions with. However, I was able to get a reasonably performant random forests model which showed me which variables were more or less important and in what direction some of them had tendencies towards. This would not have been possible with purely focusing on correlation coefficients, and this proved to me that training AI models on predictions and performing factor analysis can identify hidden results in complex data that aren't easily found by statistucal methods.

What was super interesting was how different the normal df and cdf were from df3 and cdf3. Cutting out those extreme outliers made a super significant difference in the models' prediction abilities, despite cutting out only a small portion of respondents. I tentatively conclude that it's hard to predict the students who are most drastically impacted by AI.

It seems that a student's use cases weren't nearly as impactful as I expected, to the point where I can't even make significant conclusions about whether certain use cases had noteworthy impacts on a student's perceived grade. Daily usage hours was important, and shows a slight positive tendency that aligns with my hypothesis, but it's not much to go off of.

Perhaps the most interesting conclusions with the most notable and consistent impact on the data are as follows: Major is important for a student's perception of AI's impact on their grades, and students with professors who allow AI tend to believe AI positively impacts their grade.

This dataset feels very limited by the notion of "perceived impact" as the value to predict, since a wide variety of factors could contribute to this, and my interest in the productivity of AI would be much better served by a dataset that's actually able to track the impact.

In fact, late into the project I started to suspect that much of this dataset is generated. It took me a while to notice that the Kaggle description for [this dataset](https://www.kaggle.com/datasets/rakeshkapilavai/ai-tool-usage-by-indian-college-students-2025) discussed ~500 respondents, yet the data (which has never been updated) had ~3500 rows. Additionally, the description describes "Impact on Grades" to be a value between -3 and 3, but when I restructured my project around multi-dataframe error analysis, I realized that there are some -5s, -4s, 4s, and 5s in the data (which, when removed from the data, drastically improved model performance). This could mess up my results in two ways. If I'm training to make real predictions on false data, then relationships between variables that exist in the real world but not in the generated data are largely undermined, even if the false data is generated based on some 500 real responses. On the other hand, this could mean the training data and test data sharing a relationship to some initial 500 datapoints that contributes to unidentifiable overfitting to those initial 500 points, creating improvements in model performance that shouldn't exist. I'm not 100% sure its fake, and I'm not sure training on 500 real responses versus 500 real + 3000 generated would've yielded different results, but I was too late into the project to pivot datasets. In the future, I would likely use a dataset created for another study (maybe one more focused on statistical methods), and apply machine learning models as a novel method to see if it can discover new relationships.
