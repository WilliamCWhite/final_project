### 1.2 Basic Data Exploration

Just from looking at the first 5 rows, we can get a sense of the type of data we have.  There's a `mode` column for the heavy (dedicated rail underground) and light (surface) rail identifier, route name, the day type, etc.

A full breakdown of the columns and their meaning can be found with the dataset description [here](https://gis.data.mass.gov/datasets/MassDOT::fall-2024-mbta-rail-ridership-by-sdp-time-period-route-line-and-stop/about). 

**Question**: What do `total_ons` and `total_offs` mean? How are the `average_ons` and `average_offs` determined? 

We can then explore some basic characteristics of the data, including the shape of the dataset and basic statistics about the columns. This basic exploration allows us to spot potential outliers and mistakes in the data.

```python
print(df.shape)
```

The data has 2728 rows and 18 columns.

```python
print(df.describe())
```

We notice that there is a stop with no rides (minimum is zero for `total_ons`). Also, the stop with the maximum `total_ons` to be about 20x the mean. Let's have a look at what stop that is.

Fortunately, Pandas allows to select a single column from the data by indexing using the name of the column. For example `df['total_ons']` will give us just the `total_ons` column.

```python
# Find the stop with the most rides using Boolean array indexing. 
max_row = df[df['total_ons'] == df['total_ons'].max()]
print(max_row['route_name'].tolist(), max_row['stop_name'].tolist())
```

Let's find the 0 `total_ons` stops.

```python
zero = df[df['total_ons'] == 0]

# display 10 of them
print(zero['route_name'].head())
```

We can group by columns and types to get a better understanding of what might be going on. We then use these groups to count how many times a stop has zero people getting on by day type (ignoring time of day for now). 

```python
zerogroups = zero.groupby(['route_name', 'stop_name','day_type_name']).count()
# Remove redundant colums and rename to something sensible
zerogroups = zerogroups[['mode']].rename(columns = {'mode': 'count'})

print(zerogroups)
```

**Note:** The groupby() function is very powerful, enabling statistics and feature extraction from combinations (groups) of data attributes. However, groupby() does not return an array directly, but a GroupBy object that must have other functions applied to it to be useful. In the example above, we are applying `.count()` to count the number of days with zero rides for each stopname/daytype pair.

# 2. Plotting Data

Plotting data is extremely useful for understanding it's properties. We will be using a combination of the Matplotlib, Seaborn, and Pandas libraries for plotting data in this class. With plots, your aim should be to make them clear, nondeceptive, and well-labeled. 

### 2.1 A Basic Plot


```python
# import libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Tell Jupyter notebooks to put the plot below the cell rather than in a separate window
%matplotlib inline 

# set the size of the plot
sns.set(rc={'figure.figsize':(20, 4)})
```

The following example shows one way to plot the MBTA ridership from the Davis Square station by time of day:

```python
# select rides from the Davis Squre Red Line stop heading southbound during the week 
davis_square = df[(df['stop_name'] == 'Davis') & (df['dir_id'] == 'SB') & (df['day_type_name'] == 'Weekday')]

sns.barplot(davis_square, x="time_period_name", y="average_ons")

# Label the plot
plt.ylabel("Average Number On")
plt.xlabel("Time of Day")
plt.title('Davis Square Red Line (Weekday South Bound)', fontsize=14)
plt.show()
# There are many more matplotlib functions for making plots prettier. 
# The Matplotlib documentation has examples of their use, and we will see more in future class exercises
```

### 2.2 A More Complicated Plot

The above plot is realtively simple and doesn't make full use of some of the nice ways to plot data. Below walks through a more involved plot. Our aim is to see how rides are distributed over time across lines with differing use and to see how daytype interacts with ridership. 

#### 2.2.1 Custom Groupby and Adding New Column


Wouldn't it be cool to plot ridership across the years by line. However, there's a small problem. 

```python
len(df['stop_name'].unique())
```

As you can see, there are a crazy number of lines. Something like 4 groups would be easier to plot, right? To do that, we need to bin our data. Let's first copy our data so we don't change the DataFrame we already loaded. We will group by station, sum the rides, and apply qcut to bin the data into four equally sized groups. 

```python
custom = df.copy()
groups = custom.groupby(['stop_name'])['average_ons'].sum().to_frame().apply(lambda x: pd.qcut(x, 4, labels=range(1,5))).reset_index()
```

Let's see if we successfully made equally sized groups: 

```python
groups.groupby(['average_ons']).count()
```

Let's add this label to our larger data using map. 

```python
mapping = dict(groups.to_records(index=False))
custom['freq'] = custom['stop_name'].map(lambda x: mapping[x])
```

#### 2.2.1 Facet Plots


Let's plot our data by grouping by frequency bin. We will use catplot, but FacetGrid is the more generic way of achieiving this. We will first aggregate our data by frequency bin. Aggregrate allows us to do custom operations to columns in our dataframe. Here we take the mean of the average on and take the first stationname for the relevant grouping as the stationname.

```python
plot_data = custom.groupby(['freq', 'dir_id', 'day_type_name']).agg({'average_ons': 'mean', 
                                                     'stop_name': 'first'}).reset_index()
```

```python
plot_data.head()
```

```python
sns.set(font_scale=2)

g = sns.catplot(
    data=plot_data, x="day_type_name", y="average_ons",
    col="freq",  kind="bar"
)
g.set_ylabels("Average Number On")
g.set_xlabels("Day of Week")
g.set_titles("Frequency {col_name}")
g.fig.suptitle("Ridership on Least (1) to Most (4) Number of Riders On", y=1.1)
plt.show()
```

It's a bit hard to see things when they are all on the same row. Let's use col_wrap to change up the layout.

```python
sns.set(font_scale=1.2)

g = sns.catplot(
    data=plot_data, x="day_type_name", y="average_ons",
    col="freq",  kind="bar", 
    col_wrap=2
)
g.set_ylabels("Average Number On")
g.set_xlabels("Day of Week")
g.set_titles("Frequency {col_name}")
g.fig.suptitle("Ridership on Least (1) to Most (4) Number of Riders On", y=1.1)
plt.show()
```

#### 2.2.1 Using Hue


Sometimes, we may want to include a contrast in our plot. Let's try that by focusing in on the most frequent bin of stations and see how direction may influence ridership.

```python
g = sns.barplot(
    data=plot_data[plot_data['freq'] == 4], x="day_type_name", y="average_ons",
    hue='dir_id'
)

g.set(xlabel='Day of Week', ylabel='Average Number On', title="Ridership on the Most Frequent Lines Grouped by Direction")
plt.show()
```

You can even combine these plot features!

#### 3.5 Data Exploration on Your Own

Pick a simple question or hypothesis related to the ridership data, justify **why** you picked that question (i.e., why it might be an interesting question to some audience, such as city officials), and present a simple analysis including a plot. 

I want to find the 5 stops with the highest average flow on saturdays and compare that with the average flow of those stops on weekdays. By comparing the shape of these two bar charts, city officials could see whether certain places within this group are more "weekend oriented" than others.

I would put these in a hue separated single bar chart, but I couldn't figure out how.

```python
# YOUR CODE HERE (add cells below as needed)
saturday_df = df[(df['day_type_name'] == 'Saturday')]
groups4 = saturday_df.groupby(['stop_name'])['total_ons'].sum().nlargest(5).sort_index(ascending=False)
new_sat_df = groups4.to_frame()
stops = groups4.keys()

sns.barplot(new_sat_df, x="stop_name", y="total_ons")

# Label the plot
plt.ylabel("Total Ons")
plt.xlabel("Stop Name")
plt.title('Saturday Top Stops: Total Ons on Saturdays', fontsize=14)
plt.show()

weekday_df = df[(df['day_type_name'] == 'Weekday') & (df['stop_name'].isin(stops))]
groups5 = weekday_df.groupby(['stop_name'])['total_ons'].sum().apply(lambda x: x / 5).sort_index(ascending=False)
new_weekday_df = groups5.to_frame()

sns.barplot(new_weekday_df, x="stop_name", y="total_ons")

# Label the plot
plt.ylabel("Total Ons")
plt.xlabel("Stop Name")
plt.title('Saturday Top Stops: Total Ons on Weekdays', fontsize=14)
plt.show()

# print(stops)
# print(groups4)
# print(groups5)
```

For an example of a takeaway, you'll notice that the ratio of ons between Government Center and Downtown Crossing on Weekdays is greater than that ratio on Saturdays. This likely means Government Center is relatively more active than Downtown Crossing on weekdays and relatively less on weekends.
